```python
%matplotlib inline
import math
import random
import statistics as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
class Game:
    def __init__(self):
        self.score_list = []

    def play_again(self):
        self.again = input("Would you like to run again? y/N")
        return self.again == 'y'
```

```python
class Player:
    def __init__(self, limit = 1, turns_max=7):
        self.limit = limit
        self.turns_max = turns_max
        self.score = 0
        self.turn_count = 0
        self.hold = False

    def run_round(self):
        self.reset()
        while self.turn_count < self.turns_max:
            self.turn_count += 1
            self.turn_total = self.run_turn()
            self.update_score(self.score, self.turn_total)
        return self.round_over()

    def get_roll(self):
        roll = random.randint(1,6)
        return roll

    def run_turn(self):
        self.hold = False
        self.turn_rolls = 0
        while not self.hold and self.turn_rolls < self.limit:
            self.roll = self.get_roll()
            if self.roll == 1:
                return 0
            self.turn_rolls += self.roll
            self.hold = self.will_hold()

        return self.turn_rolls

    def reset(self):
        self.score = 0
        self.turn_count = 0
        self.hold = False

    def will_hold(self):
        return self.score >= self.limit

    def update_score(self,score ,roll_total):
        self.score += roll_total
        return self.score

    def round_over(self):
        return self.score
```

```python
class RandomHoldPlayer(Player):
    def __init__(self):
        super().__init__(limit = 100)

    def will_hold(self):
        return random.choice([True, False])
```

```python
class RandomLimitPlayer(Player):
    def _init__(self, **kwargs):
        super().__init__(**kwargs)
        self.limit = random.randint(5,15)

    def round_over(self):
        self.limit = random.randint(5,15)
        return self.score
```

```python
class CautiousPlayer(Player):
    def _init__(self, **kwargs):
        super().__init__(**kwargs)
        self.limit = random.randint(3,7)

    def round_over(self):
        self.limit = random.randint(3, 7)
        return self.score
```

```python
class NormalPlayer(Player):
    def _init__(self, **kwargs):
        super().__init__(**kwargs)
        self.limit = random.randint(5,15)

    def round_over(self):
        self.limit = random.randint(5, 15)
        return self.score
```

```python
class AgressivePlayer(Player):
    def _init__(self, **kwargs):
        super().__init__(**kwargs)
        self.limit = random.randint(20,25)

    def round_over(self):
        self.limit = random.randint(20, 25)
        return self.score
```

```python
def play(who):

    game = Game()
    if who == 1:
        player = Player()
    elif who == 2:
        player = RandomHoldPlayer()
    elif who == 3:
        player = RandomLimitPlayer(random.randint(5,15))
    elif who == 4:
        player = CautiousPlayer(random.randint(3,7))
    elif who == 5:
        player = NormalPlayer(random.randint(5, 15))
    elif who == 6:
        player = AgressivePlayer(random.randint(15,25))

    for _ in range(1,101):

        score = player.run_round()
        game.score_list.append(score)

#         if not game.play_again():
#             break

    print("These were your scores for each game: ", game.score_list)
    return game.score_list
```

```python
player1_scores = play(1)
```

```python
plt.hist(player1_scores)

mean = st.mean(player1_scores)
stdev = st.stdev(player1_scores)
ymin, ymax = plt.ylim()

plt.vlines(mean, ymin, ymax, color='red')
plt.vlines([mean - 2 * stdev, mean - stdev, mean + stdev, mean + 2 * stdev], ymin, ymax, color='red', linestyles='dashed')
plt.show()
```

```python
print("\nPlayer 1 mean: ", st.mean(player1_scores))
print("Player 1 stdev: ", st.stdev(player1_scores))

print("\nPlayer 2 mean: ", st.mean(player2_scores))
print("Player 2 stdev: ", st.stdev(player2_scores))


print("\nPlayer 3 mean: ", st.mean(player3_scores))
print("Player 3 stdev: ", st.stdev(player3_scores))

print("\nPlayer 4 mean: ", st.mean(player4_scores))
print("Player 4 stdev: ", st.stdev(player4_scores))

print("\nPlayer 5 mean: ", st.mean(player5_scores))
print("Player 5 stdev: ", st.stdev(player5_scores))

print("\nPlayer 6 mean: ", st.mean(player6_scores))
print("Player 6 stdev: ", st.stdev(player6_scores))
```

# Player 6, the aggressive one, has the best chances. They have the highest average and a higher than average deviation. The highes outweigh the lows in the end. Player 2, who randomly chooses to roll or hold, was close behind, but had a wider spread. It seems that the old expression "Go big or go home" applies here. Increasing the bounds of the roll total for the aggressive player to hold at makes the mean drop and stdev increase fairly quickly. Holding somewhere between 15-30 seems to be best.
